## Actividad No. 5. Optimización por enjambre de particulas (PSO, por sus siglas en ingles, Particle Swarm Optimization) Utilizando la Librería pyswarm.

## PSO es un algoritmo de optimización basado en la simulación de la conducta social de organismos, como bandadas de aves o bancos de peces. En esta práctica implementaremos PSO para buscar el optimo global de una funcion en 2D (minimo y maximo global) y la graficamos: (f(x) = -(1 - ((11/2) * x - 7/2)**2) * (np.cos((11/2)* x - 7/2) + 1) + 2

## Paso 1: Importar librerías necesarias


```python
import numpy as np
import matplotlib.pyplot as plt 
from pyswarm import pso
```

## Paso 2: Definir función objetivo. 

```python
# Definir la función a optimizar (maximización)
def func(x):
    return -(1 - ((11/2) * x - 7/2)**2) * (np.cos((11/2)* x - 7/2) + 1) + 2
```

## Paso 3: Definir los limites de búsqueda.

```python
lb = [-10] # Limite inferior
ub = [10] # Limite superior
```

## Paso 4: Ejecutar PSO.

```python
# Ejecutar el PSO
x_opt, f_opt = pso(func, lb, ub, swarmsize = 50, maxiter = 100)

# Mostrar los resultados
print("La solución óptima es x =", x_opt)
print("El valor óptimo de la funcion es f(x) =", -f_opt)
```

## Paso 5: Graficar la función y la solución óptima (MINIMO DE LA FUNCIÓN)

```python
# Graficar la función y la solución óptima
x = np.linspace(-10, 10, 400)
# Maximo
# y = (1 - ((11/2) * x - 7/2)**2) * (np.cos((11/2)* x - 7/2) + 1) + 2
# Minimo
y = -(1 - ((11/2) * x - 7/2)**2) * (np.cos((11/2)* x - 7/2) + 1) + 2
plt.figure(figsize = (10, 6))
plt.plot(x, y, label = "f(x)")
plt.plot(x_opt, -f_opt, "ro", label = "Minimo encontrado (PSO)")
plt.title("Optimización de f(x) = -(1 - ((11/2) * x - 7/2)**2) * (np.cos((11/2)* 2 - 7/2) + 1) + 2. usando PSO")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show
```

## Paso 6: Graficar la función y la solución óptima (MAXIMO DE LA FUNCIÓN)

```python
# Graficar la función y la solución óptima
x = np.linspace(-10, 10, 400)
# Maximo
y = (1 - ((11/2) * x - 7/2)**2) * (np.cos((11/2)* x - 7/2) + 1) + 2
# Minimo
#y = -(1 - ((11/2) * x - 7/2)**2) * (np.cos((11/2)* x - 7/2) + 1) + 2
plt.figure(figsize = (10, 6))
plt.plot(x, y, label = "f(x)")
plt.plot(x_opt, -f_opt, "ro", label = "Minimo encontrado (PSO)")
plt.title("Optimización de f(x) = -(1 - ((11/2) * x - 7/2)**2) * (np.cos((11/2)* 2 - 7/2) + 1) + 2. usando PSO")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show
```
