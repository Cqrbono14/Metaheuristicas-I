## Actividad No. 1. Ascenso y Descenso del Gradiente en 2D.

### - Definimos el error que comete una red neuronal como la diferencia entre el resultado esperado en una instancia de entrenamiento y el resultado obtenido.
### - La idea principal es representar el error como una función continua de los pesos e intentar llevar la red hacia una configuracion de valores de activacion que nos permita encontrar el mínimo de esta función.
### - Para determinar la magnitud y dirección del cambio debemos introducir en el vector de pesos para reducir el error que está cometiendo, utilizaremos un concepto de "pendiente", que indica que la pendiente de un punto dado es el gardiente de la tangente a la curva de la función en el punto dado
### - El gradiente es la derivada de una operacion tensorial. Es la generalización del concepto de derivadas a funciones que toman tensores como entradas.

## Paso 1: Importar librerías necesarias


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sympy import Symbol, cos, diff, lambdify
# from sklearn.model_selection import train_test_split
```

## Paso 2: Definimos la función: f(x) = 2x^2 * cos(x) - 5x

```python
x = Symbol('x')

f = 2 * x**2 * cos(x) - 5 * x
```

## Paso 3: Graficamos la función


```python
figsize = (6, 4)
dpi = 150

fig, ax = plt.subplots(1, 1, figsize = figsize, dpi = dpi)

t1 = np.arange(-5, 5, 0.1)

# Convertir la función a funcion numérica
f_func = lambdify(x, f, 'numpy')

ax.set_xlim(-5, 5)
ax.set_ylim(-45, 45)
ax.plot(t1, f_func(t1))
```

# El Algoritmo del Descenso del Gradiente


## Paso 3.1: Tome un punto al azar X0

```python
fig, ax = plt.subplots(1, 1, figsize = figsize, dpi = dpi)

t1 = np.arange(-5, 5, 0.1)

ax.set_xlim(-5, 5)
ax.set_ylim(-45, 45)
ax.plot(t1, f_func(t1))

x_value = [-1]
y_value = [float(f_func(x_value[0]))]
ax.plot(x_value, y_value, 'ro')

print("X = ", x_value)
print("Y = ", y_value)
```

## Paso 3.2: Calcule el valor de la pendiente (derivada) f'(X0)

```python
# La derivada de f (La pendiente)
df = diff(f, x)
```

```python
# Convertir la derivada de la función a funcion numérica
df_func = lambdify(x, df, 'numpy')

pendiente = df_func(x_value[0])
print("Pendiente: ", pendiente)
```

## Paso 3.3: Camine en dirección opuesta a la pendiente: x1=x0 - eta * f'(x0)
## Nota. Aquí eta es la tasa de aprendizaje que mencionamos previamente.

```python
eta = 0.05
#eta = 0.7
x_value.append(float(x_value[0] - eta * pendiente))
y_value.append(float(f_func(x_value[0] - eta * pendiente)))
print("X = ", x_value)
print("Y = ", y_value)
```

## Gráficamos el punto 2.

```python
fig, ax = plt.subplots(1, 1, figsize = figsize, dpi = dpi)
ax.set_xlim(-5, 5)
ax.set_ylim(-45, 45)
ax.plot(t1, f_func(t1))
ax.plot(x_value, y_value, 'ro')
```

## Gráficamos un tercer punto

```python
x_value.append(float(x_value[1] - eta * df_func(x_value[1])))
y_value.append(float(f_func(x_value[1] - eta * df_func(x_value[1]))))
print("X = ", x_value)
print("Y = ", y_value)

pendiente = df_func(x_value[0])
print("Pendiente:x[0]: ", pendiente)
pendiente = df_func(x_value[1])
print("Pendiente:x[1]: ", pendiente)
pendiente = df_func(x_value[2])
print("Pendiente:x[2]: ", pendiente)

fig, ax = plt.subplots(1, 1, figsize = figsize, dpi = dpi)
ax.set_xlim(-5, 5)
ax.set_ylim(-45, 45)
ax.plot(t1, f_func(t1))
ax.plot(x_value, y_value, 'ro')
```

## Graficamos 20 puntos mas

```python
eta = 0.05
#eta = 0.07
x_value = [-1.0]
y_value = [f_func(x_value[0])]
pendiente_value = [df_func(x_value[0])]
for i in range(20):
    x_value.append(x_value[i] - eta * df_func(x_value[i]))
    y_value.append(f_func(x_value[i] - eta * df_func(x_value[i])))
    pendiente_value.append(df_func(x_value[i]))

dt = pd.DataFrame({"X": x_value, "Y": y_value, "Pendiente": pendiente_value})
print(dt)

fig, ax = plt.subplots(1, 1, figsize = figsize, dpi = dpi)
ax.set_xlim(-5, 5)
ax.set_ylim(-45, 45)
ax.plot(t1, f_func(t1))
ax.plot(x_value, y_value, 'ro')
```

# El Algoritmo del Ascenso del Gradiente


## Paso 3.4: Tome un punto al azar X0

```python
fig, ax = plt.subplots(1, 1, figsize = figsize, dpi = dpi)

t1 = np.arange(-5, 5, 0.1)

ax.set_xlim(-5, 5)
ax.set_ylim(-45, 45)
ax.plot(t1, f_func(t1))

x_value = [2]
y_value = [float(f_func(x_value[0]))]
ax.plot(x_value, y_value, 'ro')

print("X = ", x_value)
print("Y = ", y_value)
```

## Paso 3.5: Calcule el valor de la pendiente (derivada) f'(X0)

```python
# La derivada de f (La pendiente)
df = diff(f, x)
```

```python
# Convertir la derivada de la función a funcion numérica
df_func = lambdify(x, df, 'numpy')

pendiente = df_func(x_value[0])
print("Pendiente: ", pendiente)
```

## Paso 3.6: Camine hacia la pendiente: x1=x0 + eta * f'(x0)
## Nota. Aquí eta es la tasa de aprendizaje que mencionamos previamente.

```python
eta = 0.05
#eta = 0.7
x_value.append(float(x_value[0] + eta * pendiente))
y_value.append(float(f_func(x_value[0] + eta * pendiente)))
print("X = ", x_value)
print("Y = ", y_value)
```

## Gráficamos el punto 2.

```python
fig, ax = plt.subplots(1, 1, figsize=figsize, dpi=dpi)
ax.set_xlim(-5, 5)
ax.set_ylim(-45, 45)
ax.plot(t1, f_func(t1))
ax.plot(x_value, y_value, 'ro')
```

## Gráficamos un tercer punto

```python
x_value.append(float(x_value[1] + eta * df_func(x_value[1])))
y_value.append(float(f_func(x_value[1] + eta * df_func(x_value[1]))))
print("X = ", x_value)
print("Y = ", y_value)

pendiente = df_func(x_value[0])
print("Pendiente:x[0]: ", pendiente)
pendiente = df_func(x_value[1])
print("Pendiente:x[1]: ", pendiente)
pendiente = df_func(x_value[2])
print("Pendiente:x[2]: ", pendiente)

fig, ax = plt.subplots(1, 1, figsize=figsize, dpi=dpi)
ax.set_xlim(-5, 5)
ax.set_ylim(-45, 45)
ax.plot(t1, f_func(t1))
ax.plot(x_value, y_value, 'ro')
```

## Graficamos 20 puntos mas

```python
eta = 0.05
#eta = 0.07
x_value = [2]
y_value = [f_func(x_value[0])]
pendiente_value = [df_func(x_value[0])]
for i in range(20):
    x_value.append(x_value[i] + eta * df_func(x_value[i]))
    y_value.append(f_func(x_value[i] + eta * df_func(x_value[i])))
    pendiente_value.append(df_func(x_value[i]))

dt = pd.DataFrame({"X": x_value, "Y": y_value, "Pendiente": pendiente_value})
print(dt)

fig, ax = plt.subplots(1, 1, figsize = figsize, dpi = dpi)
ax.set_xlim(-5, 5)
ax.set_ylim(-45, 45)
ax.plot(t1, f_func(t1))
ax.plot(x_value, y_value, 'ro')
```
