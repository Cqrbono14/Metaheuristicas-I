## Actividad No. 6. Optimización por enjambre de particulas (PSO, por sus siglas en ingles, Particle Swarm Optimization) Utilizando la Librería pyswarm.

## PSO es un algoritmo de optimización basado en la simulación de la conducta social de organismos, como bandadas de aves o bancos de peces. En esta práctica implementaremos PSO para buscar el optimo global de una funcion en 3D (minimo global): f(x, y) = x^2 + y^2 y la graficamos

## Paso 1: Importar librerías necesarias.

```python
from pyswarm import pso
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Paso 2: Definir función objetivo. 

```python
# Definir la función objetivo: f(x, y) = x^2 + y^2
def objetive_function(x):
    # x es un array de la forma [x, y]
    return x[0]**2 + x[1]**2
```

## Paso 3: Definir los limites de búsqueda.

```python
# Definir los limites de busqueda para x e y
# Supongamos que queremos que x e y esten entre -10 y 10
lb = [-10, -10] # Limites inferiores
ub = [10, 10] # Limites superiores
```

## Paso 4: Ejecutar PSO.

```python
# Ejecutar el algoritmo PSO usando pyswarm
best_position, best_value = pso(objetive_function, lb, ub, swarmsize = 30, maxiter = 100)

# Imprimir los resultados 
print(f"Mejor posición encontrada: {best_position}")
print(f"Valor de la función objetivo en la mejor posición: {best_value}")
```

## Paso 5: Graficar la función y los resultados de PSO en 3D (MINIMO DE LA FUNCIÓN)

```python
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize = (12, 8))
ax = fig.add_subplot(111, projection = "3d")
ax.plot_surface(X, Y, Z, cmap = "viridis", alpha = 0.7)
ax.scatter(best_position[0], best_position[1], best_value, color = "red", s = 100, label = "Mejor Posición Encontrada")
```
